# Dữ liệu thực phẩm USDA

### Import các thư viện cần thiết và đọc dữ liệu

```python
import json
import pandas as pd

db = json.load(open("datasets/usda_food/database.json"))
len(db)
```

### Xem các cấu trúc dữ liệu

```python
db[0].keys()
```

```python
db[0]["nutrients"][0]
```

```python
nutrients = pd.DataFrame(db[0]["nutrients"])
nutrients.head(7)
```

```python
info_keys = ["description", "group", "id", "manufacturer"]
info = pd.DataFrame(db, columns=info_keys)
info.head()
```

```python
info.info()
```

```python
pd.value_counts(info["group"])[:10]
```

### Làm sạch và tiền xử lý
- Xử lý NA/giá trị bất thường, đổi tên cột, chuẩn hóa kiểu dữ liệu.
- Gộp/ghép bảng, lọc dòng/cột cần thiết.

```python
nutrients = []
for rec in db:
    fnuts = pd.DataFrame(rec["nutrients"])
    fnuts["id"] = rec["id"]
    nutrients.append(fnuts)
    
nutrients = pd.concat(nutrients, ignore_index=True)

nutrients
```

```python
nutrients.duplicated().sum() # number of duplicates
```

```python
nutrients = nutrients.drop_duplicates()
```

```python
col_mapping = {"description" : "food", "group" : "fgroup"}
info = info.rename(columns=col_mapping, copy=False)
info.info()
```

```python
col_mapping = {"description" : "nutrient", "group" : "nutgroup"}
nutrients = nutrients.rename(columns=col_mapping, copy=False)
nutrients
```

### Tiến hành EDA và trực quan hoá
- Gộp bảng và phân tích

```python
ndata = pd.merge(nutrients, info, on="id")
ndata.info()
```

```python
ndata.iloc[30000]
```

- Tạo biểu đồ/hệ số để hiểu phân phối và mối quan hệ.

```python
result = ndata.groupby(["nutrient", "fgroup"])["value"].quantile(0.5)
result["Zinc, Zn"].sort_values().plot(kind="barh")
```

- Tìm ra loại thực phẩm chứa nhiều nhất một chất dinh dưỡng cụ thể.

```python
by_nutrient = ndata.groupby(["nutgroup", "nutrient"])
def get_maximum(x):
    return x.loc[x.value.idxmax()]
max_foods = by_nutrient.apply(get_maximum)[["value", "food"]]
# make the food a little smaller
max_foods["food"] = max_foods["food"].str[:50]
```
