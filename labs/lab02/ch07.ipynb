```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

Đoạn này tạo một DataFrame mới với hai cột là loại thực phẩm ("food") và trọng lượng ("ounces").

```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
```

```python
data["animal"] = data["food"].map(meat_to_animal)
data
```

Đoạn này sử dụng phương thức .map() của Series trên cột "food" và áp dụng dictionary meat_to_animal để tạo ra một cột mới ("animal") chứa tên loài động vật tương ứng với mỗi loại thực phẩm.

```python
def get_animal(x):
    return meat_to_animal[x]
data["food"].map(get_animal)
```

Đoạn này định nghĩa một hàm Python đơn giản (get_animal) để tra cứu giá trị trong dictionary, sau đó sử dụng .map() để áp dụng hàm này lên cột "food". Kết quả tương tự như việc truyền thẳng dictionary vào .map().

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

 Đoạn này tạo một Series số, trong đó các giá trị đặc biệt như -999 và -1000 được coi là các giá trị bị thiếu (hoặc bất thường) cần được thay thế.

```python
data.replace(-999, np.nan)
```

Đoạn này sử dụng phương thức .replace() để thay thế tất cả các lần xuất hiện của giá trị -999 bằng giá trị NaN (dữ liệu bị thiếu).

```python
data.replace([-999, -1000], np.nan)
```

Đoạn này sử dụng .replace() với một list các giá trị cần thay thế ([-999, -1000]) và thay thế tất cả chúng bằng một giá trị duy nhất là np.nan.

```python
data.replace([-999, -1000], [np.nan, 0])
```

Đoạn này sử dụng .replace() với hai list: list đầu tiên là các giá trị cần thay thế ([-999, -1000]), và list thứ hai là các giá trị sẽ được thay thế tương ứng ([np.nan, 0]). Cụ thể, -999 được thay bằng np.nan và -1000 được thay bằng 0.

```python
data.replace({-999: np.nan, -1000: 0})
```

Đoạn này thực hiện việc thay thế tương tự như đoạn trước, nhưng sử dụng một dictionary để chỉ định rõ ràng cặp giá trị cần thay thế và giá trị thay thế.

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=["Ohio", "Colorado", "New York"],
                    columns=["one", "two", "three", "four"])
```

```python
def transform(x):
    return x[:4].upper()

data.index.map(transform)
```

Đoạn này định nghĩa một hàm (transform) để lấy 4 ký tự đầu tiên của một chuỗi và chuyển chúng thành chữ hoa. Sau đó, nó sử dụng .map() để áp dụng hàm này lên các nhãn hàng (index) của DataFrame.

```python
data.index = data.index.map(transform)
data
```

Đoạn này gán kết quả của phép biến đổi (từ .map(transform)) trở lại làm nhãn hàng của DataFrame, thay đổi nhãn hàng từ "Ohio", "Colorado", "New York" thành "OHIO", "COLO", "NEW ".

```python
data.rename(index=str.title, columns=str.upper)
```

Đoạn này sử dụng phương thức .rename() để đổi tên các nhãn hàng và cột.

index=str.title: Chuyển nhãn hàng thành định dạng Title Case (ví dụ: "OHIO" thành "Ohio").

columns=str.upper: Chuyển nhãn cột thành chữ hoa (ví dụ: "one" thành "ONE"). Lưu ý: rename() tạo ra một đối tượng mới, không thay đổi trực tiếp data.

```python
data.rename(index={"OHIO": "INDIANA"},
            columns={"three": "peekaboo"})
```

Đoạn này sử dụng .rename() với các dictionary để đổi tên cụ thể một nhãn hàng ("OHIO" thành "INDIANA") và một nhãn cột ("three" thành "peekaboo").

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)
age_categories
```

Đoạn này sử dụng hàm pd.cut() để phân nhóm (rời rạc hóa) list ages thành các khoảng dựa trên ranh giới được định nghĩa trong list bins (ví dụ: (18, 25], (25, 35],...). Kết quả là một đối tượng Categorical.

```python
age_categories.codes
age_categories.categories
age_categories.categories[0]
pd.value_counts(age_categories)
```

```python
pd.cut(ages, bins, right=False)
```

Đoạn này sử dụng pd.cut() với tham số right=False, nghĩa là các khoảng phân nhóm bây giờ sẽ bao gồm cận dưới và loại trừ cận trên (ví dụ: [18, 25), [25, 35),...).

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]
pd.cut(ages, bins, labels=group_names)
```

 Đoạn này sử dụng tham số labels trong pd.cut() để gán tên tùy chỉnh cho các khoảng phân nhóm thay vì sử dụng tên khoảng mặc định.

```python
data = np.random.uniform(size=20)
pd.cut(data, 4, precision=2)
```

 Đoạn này tạo ra 20 số ngẫu nhiên phân phối đều (uniform), sau đó sử dụng pd.cut() để chia dữ liệu thành 4 khoảng có độ dài bằng nhau (dựa trên giá trị tối thiểu và tối đa của dữ liệu), và hiển thị ranh giới khoảng với độ chính xác là 2 chữ số thập phân.

```python
data = np.random.standard_normal(1000)
quartiles = pd.qcut(data, 4, precision=2)
quartiles
pd.value_counts(quartiles)
```

 Đoạn này tạo 1000 số ngẫu nhiên phân phối chuẩn. Hàm pd.qcut() được sử dụng để phân nhóm dữ liệu dựa trên tứ phân vị (quantile), đảm bảo rằng mỗi nhóm sẽ có xấp xỉ số lượng điểm dữ liệu bằng nhau. pd.value_counts() xác nhận điều này.

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]).value_counts()
```

Đoạn này sử dụng pd.qcut() với các ngưỡng phân vị tùy chỉnh (0%, 10%, 50%, 90%, 100%) để chia dữ liệu thành các nhóm không đều về kích thước, nhưng được xác định dựa trên tỷ lệ phần trăm của dữ liệu.

```python
df = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                   "data1": range(6)})
df
pd.get_dummies(df["key"], dtype=float)
```

Đoạn này tạo một DataFrame. Sau đó, nó sử dụng hàm pd.get_dummies() trên cột "key" để chuyển đổi các biến phân loại (categorical variables) thành các biến giả/biến chỉ thị (dummy/indicator variables) (còn gọi là One-Hot Encoding).

```python
dummies = pd.get_dummies(df["key"], prefix="key", dtype=float)
df_with_dummy = df[["data1"]].join(dummies)
df_with_dummy
```

Đoạn này tạo các biến giả từ cột "key" với tiền tố là "key_" và sau đó sử dụng phương thức .join() để kết hợp các cột biến giả này vào DataFrame gốc (chỉ giữ lại cột "data1" từ DataFrame gốc).

```python
mnames = ["movie_id", "title", "genres"]
movies = pd.read_table("datasets/movielens/movies.dat", sep="::",
                       header=None, names=mnames, engine="python")
movies[:10]
```

Đoạn này đọc dữ liệu từ tệp movies.dat, sử dụng :: làm ký tự phân cách, gán tên cột và in ra 10 hàng đầu tiên của DataFrame movies.

```python
dummies = movies["genres"].str.get_dummies("|")
dummies.iloc[:10, :6]
```

Đoạn này sử dụng phương thức chuỗi .str.get_dummies("|") để tạo các biến giả từ cột "genres" bằng cách phân tách chuỗi dựa trên ký tự |. Mỗi thể loại (genre) sau khi tách sẽ trở thành một cột biến giả mới (One-Hot Encoding).

```python
movies_windic = movies.join(dummies.add_prefix("Genre_"))
movies_windic.iloc[0]
```

 Đoạn này kết hợp các cột biến giả dummies (với tiền tố "Genre_" được thêm vào) với DataFrame movies gốc bằng phương thức .join() và in ra hàng đầu tiên.

```python
import re
text = "foo    bar\t baz  \tqux"
re.split(r"\s+", text)
```

Đoạn này nhập thư viện re (biểu thức chính quy). Hàm re.split(r"\s+", text) được sử dụng để tách chuỗi text dựa trên một hoặc nhiều ký tự khoảng trắng (\s+ bao gồm dấu cách, tab, newline, v.v.).

```python
regex = re.compile(r"\s+")
regex.split(text)
```

 Đoạn này biên dịch (compile) biểu thức chính quy (r"\s+") thành một đối tượng regex. Sau đó, nó sử dụng phương thức .split() của đối tượng regex đã biên dịch để tách chuỗi.



```python
regex.findall(text)
```

Đoạn này sử dụng phương thức .findall() của đối tượng regex để tìm và trả về tất cả các chuỗi con khớp với biểu thức chính quy (r"\s+"), tức là tìm ra tất cả các chuỗi khoảng trắng.

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""
pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"

# re.IGNORECASE makes the regex case insensitive
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
regex.findall(text)
```

Đoạn này sử dụng phương thức .findall() của đối tượng regex để tìm và trả về tất cả các địa chỉ email (khớp với pattern) trong chuỗi text.

```python
m = regex.search(text)
m
text[m.start():m.end()]
```

Đoạn này sử dụng phương thức .search() để tìm lần khớp đầu tiên với pattern trong chuỗi text.
m: Là đối tượng khớp (match object).
text[m.start():m.end()]: Trích xuất chuỗi con đã khớp (địa chỉ email đầu tiên) dựa trên vị trí bắt đầu và kết thúc của đối tượng khớp.



```python
print(regex.match(text))
```

Đoạn này sử dụng phương thức .match() để tìm kiếm sự khớp ngay từ đầu chuỗi. Vì chuỗi text bắt đầu bằng "Dave " chứ không phải một địa chỉ email, nên .match() không tìm thấy sự khớp và trả về None.

```python
rng = np.random.default_rng(seed=12345)
draws = rng.standard_normal(1000)
draws[:5]
```

Đoạn này tạo 1000 số ngẫu nhiên phân phối chuẩn để sử dụng cho ví dụ tiếp theo.

```python
bins = pd.qcut(draws, 4)
bins
```

 Đoạn này sử dụng hàm pd.qcut() để phân chia mảng draws thành 4 nhóm dựa trên phân vị (quartiles), tạo ra một Series với kiểu dữ liệu Categorical.

```python
bins = pd.qcut(draws, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
bins
bins.codes[:10]
```

Đoạn này gán nhãn tùy chỉnh (Q1 đến Q4) cho các nhóm phân vị. .codes[:10] hiển thị mã số nguyên (0, 1, 2, 3) tương ứng với các nhãn này.

```python
bins = pd.Series(bins, name='quartile')
results = (pd.Series(draws)
           .groupby(bins)
           .agg(['count', 'min', 'max'])
           .reset_index())
results
```

Đoạn này sử dụng các nhóm phân vị (bins) làm khóa nhóm (key) để thực hiện tổng hợp (aggregation):
Tạo Series bins có tên.
Nhóm dữ liệu draws theo bins.
Áp dụng các hàm tổng hợp count, min, max.
.reset_index() chuyển khóa nhóm thành một cột thông thường.

```python
results['quartile']
```
